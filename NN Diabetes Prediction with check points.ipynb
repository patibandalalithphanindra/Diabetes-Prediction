```python
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
data = pd.read_csv('/content/dataset.csv')
```

```python
from sklearn.model_selection import train_test_split
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=5)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(5, input_shape=(5, )))
model.add(tf.keras.layers.Dense(5, activation="relu"))
model.add(tf.keras.layers.Dense(1, activation="sigmoid"))

model.summary()
```

```python
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.callbacks import ModelCheckpoint

opt = SGD(lr=0.01, momentum=0.9)

model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(5, input_shape=(5, )))
model.add(tf.keras.layers.Dense(5, activation="relu"))
model.add(tf.keras.layers.Dense(1, activation="sigmoid"))
model.compile(optimizer=opt, metrics='accuracy', loss='mse')

checkpoint = ModelCheckpoint('/content/checkpts_1_200.hdf5', monitor='val_accuracy', verbose=0, save_best_only=True, mode='max')
callbacks_list = [checkpoint]

model.fit(x_train, y_train, batch_size=15, epochs=200, verbose=0, validation_data=(x_test, y_test), callbacks=callbacks_list)
model.load_weights("checkpts_1_200.hdf5")
_, acc = model.evaluate(x_test, y_test)
print('1:200 '+str(acc))

model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(5, input_shape=(5, )))
model.add(tf.keras.layers.Dense(5, activation="relu"))
model.add(tf.keras.layers.Dense(1, activation="sigmoid"))
model.compile(optimizer=opt, metrics='accuracy', loss='mse')

checkpoint = ModelCheckpoint('/content/checkpts_1_400.hdf5', monitor='val_accuracy', verbose=0, save_best_only=True, mode='max')
callbacks_list = [checkpoint]

model.fit(x_train, y_train, batch_size=15, epochs=400, verbose=0, validation_data=(x_test, y_test), callbacks=callbacks_list)
model.load_weights("checkpts_1_400.hdf5")
_, acc = model.evaluate(x_test, y_test)
print('1:400 '+str(acc))

model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(5, input_shape=(5, )))
model.add(tf.keras.layers.Dense(5, activation="relu"))
model.add(tf.keras.layers.Dense(1, activation="sigmoid"))
model.compile(optimizer=opt, metrics='accuracy', loss='mse')

checkpoint = ModelCheckpoint('/content/checkpts_1_800.hdf5', monitor='val_accuracy', verbose=0, save_best_only=True, mode='max')
callbacks_list = [checkpoint]

model.fit(x_train, y_train, batch_size=15, epochs=800, verbose=0, validation_data=(x_test, y_test), callbacks=callbacks_list)
model.load_weights("checkpts_1_800.hdf5")
_, acc = model.evaluate(x_test, y_test)
print('1:800 '+str(acc))
```
